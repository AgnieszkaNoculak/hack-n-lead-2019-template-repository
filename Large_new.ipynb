```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('large.csv')
```

```python
country = pd.read_csv('country.csv')
country.head()
```

```python
country_dict = country['country_name'].to_dict()
print(country_dict)
```

```python
data.head()
```

```python
data['nationality'] = data['nationality'].add(-1)
```

```python
data['nationality'].replace(country_dict, inplace = True)
data.head()
```

```python
data['nationality'].value_counts()
```

```python
data_0 = data[data['category'] == 0]
```

```python
data_1 = data[data['category'] == 1]
```

```python
data_2 = data[data['category'] == 2]
```

```python
data_0['nationality'].value_counts(sort=True, ascending=True)
```

```python
data_0['is_pep'].value_counts()
```

```python
data_0['n_of_accounts'].value_counts()
```

```python
data_0['distinct_counterparties'].value_counts()
```

```python
print(data_0['turnover'].max())
print(data_1['turnover'].max())
print(data_2['turnover'].max())

print(data_0['atm_withdrawal'].max())
print(data_1['atm_withdrawal'].max())
print(data_2['atm_withdrawal'].max())
```

```python
from matplotlib import pyplot as plt

fig, (ax1) = plt.subplots(1, 1, figsize=(15, 10))
pd.value_counts(data_0['nationality']).plot('bar', ax = ax1)
pd.value_counts(data_1['nationality']).plot('bar', ax = ax1, color = 'red')
pd.value_counts(data_2['nationality']).plot('bar', ax = ax1, color = 'black')

plt.show()
```

```python
from matplotlib import pyplot as plt

fig, ([ax1, ax2, ax3], [ax4, ax5, ax6]) = plt.subplots(2, 3, figsize=(15, 10))

pd.value_counts(data_0['nationality']).plot('bar', ax=ax6)
pd.value_counts(data_1['nationality']).plot('bar', ax=ax6, color = 'red')
pd.value_counts(data_2['nationality']).plot('bar', ax=ax6, color = 'black')
ax6.title.set_text('Nationality')
pd.value_counts(data_0['is_pep']).plot('bar', ax=ax2)
ax2.title.set_text('is_pep')
ax3.hist(data_0['inactive_days_max'], bins = 20, range = (0,200))
ax3.hist(data_1['inactive_days_max'], bins = 20, range = (0,200), color = 'red')
ax3.hist(data_2['inactive_days_max'], bins = 20, range = (0,200), color = 'black')
ax3.title.set_text('inactive_days_max')
ax4.hist(data_0['inactive_days_average'], bins = 30, range = (0,100))
ax4.hist(data_1['inactive_days_average'], bins = 30, range = (0,100), color = 'red')
ax4.hist(data_2['inactive_days_average'], bins = 30, range = (0,100), color = 'black')
ax4.title.set_text('inactive_days_average')
ax5.hist(data_0['io_ratio'], bins = 30)
ax5.hist(data_1['io_ratio'], bins = 30, color = 'red')
ax5.hist(data_2['io_ratio'], bins = 30, color = 'black')
ax5.title.set_text('io_ratio')
ax1.hist(data_0['transaction_count'], bins = 30, range = (0,150))
ax1.hist(data_1['transaction_count'], bins = 30, color = 'red', range = (0,150))
ax1.hist(data_2['transaction_count'], bins = 30, color = 'black', range = (0,150))
ax1.title.set_text('transactions_count')

plt.show()
```

```python
fig, ([ax1, ax2, ax3], [ax4, ax5, ax6]) = plt.subplots(2, 3, figsize=(15, 10))

ax1.hist(data_0['turnover'], bins = 30)
ax1.hist(data_1['turnover'], bins = 30, color = 'red')
ax1.hist(data_2['turnover'], bins = 30, color = 'black')
ax1.title.set_text('turnover')
ax2.hist(data_0['age'], bins = 20)
ax2.title.set_text('age')
pd.value_counts(data_0['n_of_accounts']).plot('bar', ax=ax3)
pd.value_counts(data_1['n_of_accounts']).plot('bar', ax=ax3, color = 'red')
pd.value_counts(data_2['n_of_accounts']).plot('bar', ax=ax3, color = 'black')
ax3.title.set_text('n_of_accounts')
ax4.hist(data_0['distinct_counterparties'], bins = 30, range = (0,50))
ax4.hist(data_1['distinct_counterparties'], bins = 30, color = 'red', range = (0,50))
ax4.hist(data_2['distinct_counterparties'], bins = 30, color = 'black', range = (0,50))
ax4.title.set_text('distinct_counterparties')
ax5.hist(data_0['channel_risk'], bins = 30)
ax5.hist(data_1['channel_risk'], bins = 30, color = 'red')
ax5.hist(data_2['channel_risk'], bins = 30, color = 'black')
ax5.title.set_text('channel_risk')
ax6.hist(data_0['atm_withdrawal'], bins = 30)
ax6.hist(data_1['atm_withdrawal'], bins = 30, color = 'red')
ax6.hist(data_2['atm_withdrawal'], bins = 30, color = 'black')
ax6.title.set_text('atm_withdrawal')

plt.show()
```
